```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from fpdf import FPDF

df = pd.read_csv('CADASTRO_IES_2020.CSV', encoding='ISO-8859-1', sep=';',low_memory=False)

#Removendo valores nan
df.dropna(inplace=True)
```

```python
#gráfico 1
cores=['blue', 'green', 'orange', 'red', 'purple']
qt_regiao = df.groupby('NO_REGIAO_IES')['QT_TEC_TOTAL'].sum()
#grafico 1
qt_regiao.plot(kind='bar', title='Quantidade de Estudantes Incritos por Região', fontsize=12.0, width=0.9, color=cores)
plt.grid(True, alpha=0.3)
plt.savefig('estudantes-regiao.png')
```

```python
#gráfico 2
fund_incom_fem = df.groupby('NO_REGIAO_IES')['QT_TEC_FUNDAMENTAL_INCOMP_FEM'].sum()
fund_incom_fem.plot(kind='bar', title='Quantidade Estudantes do Genêro Feminino com Ensino Fundamental Incompleto', fontsize=12.0, width=0.9, color=cores)
plt.grid(True, alpha=0.3)
plt.savefig('fund_incom_fem.png')

fund_incom_masc = df.groupby('NO_REGIAO_IES')['QT_TEC_FUNDAMENTAL_INCOMP_MASC'].sum()
fund_incom_masc.plot(kind='bar', title='Quantidade Estudantes do Genêro Masculino com Ensino Fundamental Incompleto', fontsize=12.0, width=0.9, color=cores)
plt.grid(True, alpha=0.3)
plt.savefig('fund_incom_masc.png')


```
