```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from fpdf import FPDF

df = pd.read_csv('CADASTRO_IES_2020.CSV', encoding='ISO-8859-1', sep=';',low_memory=False)

#Removendo valores nan
df.dropna(inplace=True)
```

```python
#gráfico 1
cores=['blue', 'green', 'orange', 'red', 'purple']
qt_regiao = df.groupby('NO_REGIAO_IES')['QT_TEC_TOTAL'].sum()
#grafico 1
qt_regiao.plot(kind='bar', title='Quantidade de Funcionários Incritos por Região', fontsize=12.0, width=0.9, color=cores)
plt.grid(True, alpha=0.3)
plt.xlabel('Regiões')
plt.savefig('funcionarios-regiao.png')
```

```python
#gráfico 2
fund_incom_fem = df.groupby('NO_REGIAO_IES')['QT_TEC_FUNDAMENTAL_INCOMP_FEM'].sum()
fund_incom_fem.plot(kind='bar', title='Quantidade Funcionários do Genêro Feminino com Ensino Fundamental Incompleto', fontsize=12.0, width=0.9, color=cores)
plt.grid(True, alpha=0.3)
plt.xlabel('Regiões')
plt.savefig('fund_incom_fem.png')

fund_incom_masc = df.groupby('NO_REGIAO_IES')['QT_TEC_FUNDAMENTAL_INCOMP_MASC'].sum()
fund_incom_masc.plot(kind='bar', title='Quantidade Funcionários do Genêro Masculino com Ensino Fundamental Incompleto', fontsize=12.0, width=0.9, color=cores)
plt.grid(True, alpha=0.3)
plt.savefig('fund_incom_masc.png')


```

```python
doc_pretos = df.groupby('NO_REGIAO_IES')['DOC_EX_PRETA'].sum()
doc_pretos.plot(kind='bar', title='Docentes Pretos por Região', fontsize=12.0, width=0.9, color=cores)
plt.grid(True, alpha=0.3)
plt.xlabel('Regiões')
plt.savefig('doc_pretos.png')
```

```python
doc_indigenas = df.groupby('NO_REGIAO_IES')['DOC_EX_INDÍGENA'].sum()
doc_indigenas.plot(kind='bar', title='Docentes Indígenas por Região', fontsize=12.0, width=0.9, color=cores)
plt.grid(True, alpha=0.3)
plt.xlabel('Regiões')
plt.savefig('indigenas.png')
```

```python
doc_total = df.groupby('NO_REGIAO_IES')['QT_DOCENTE_TOTAL'].sum()
doc_total.plot(kind='bar', title='Docentes Indígenas por Região', fontsize=12.0, width=0.9, color=cores)
plt.grid(True, alpha=0.3)
plt.xlabel('Regiões')
plt.savefig('doc_total.png')
```
